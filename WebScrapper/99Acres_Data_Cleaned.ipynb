```python
import pandas as pd
data = pd.read_csv('/content/99Acres.csv')
data.head(5)
```

```python
data.columns = ['Name', 'Builder', 'BHK', 'Price', 'Area', 'Description', 'Nearby']
```

```python
data.head(5)
```

```python
def add_dash(item):
  if '-' in item:
    return item
  else:
    return ' - ' + item
```

```python
def calculating_amount(price):
  if 'L' in price:
    return '0.' + price.replace('L', '').replace('.','').strip()
  elif 'Cr' in price:
    return price.replace('Cr', '').strip()
  else:
    return None
```

```python
data["Price in Cr"] = data["Price"].apply(add_dash).str.split('-', expand=True).get(1).str.replace('₹', '').apply(calculating_amount).astype(float)
```

```python
data['Area in sq.ft']=data['Area'].str.split(' - ',expand=True).get(1).str.replace('sq.ft', '').astype(float)
```

```python
data['Price(Rs) per sq.ft'] = round(data['Price in Cr']*10000000 / data['Area in sq.ft'], 2)
```

```python
data['BHK'] = data['BHK'].str.split(' ').str.get(0).astype(int)
```

```python
data['Bedrooms'] = data['BHK']
```

```python
data.drop(['Price','Area'], axis=1, inplace=True)
```
